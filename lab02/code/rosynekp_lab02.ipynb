In this lab, we will be exploring the steps and process of exploratory data analysis on our cleaned version of the Sacremento real estate data set that we cleaned in lab 1, that contains information on property sales in the Sacrmento area. ### **Part I - Regression on Price**

We will be exploring the continuous columns to find which variables are predictive of the price. For each continous variable, create a scatterplot of the continuous variable versus the price.

##### **Scatterplot: Price v. Square Footage**

```python
sns.scatterplot(data=df, x='sq__ft', y='price')
plt.title('Price v. Square Footage')
```

It can be observed from the scatterplot that there appears to be a linear relationship between square footage and price for properties with square footage greater than zero. For properties with a square footage of zero the relationship between the price appears to be an increasing vertical line where sq__ft equals zero. 

##### **Scatterplot: Price v. latitude**

```python
sns.scatterplot(data=df, x='latitude', y='price')
plt.title('Price v. Latitude')
```

It can be observed from the scatterplot above that there is little to no correlation between the latitude and price variable, so latititude is not predictive of price.

##### **Scatterplot: Price v. Longitude**

```python
sns.scatterplot(data=df, x='longitude', y='price')
plt.title('Price v. Longitude')
```

It can be observed from the scatterplot above that there is little correlation between the longitude and price variable, so longitude is not predictive of price.

For each categorical variable, create a box plot of the categorical variable versus price.

##### For each categorical variable, create a cross-tabulation table that shows the counts of each categorical variable value for each property type. For each categorical feature, visualize the normalized cross-tabulation table using a bar plot for each possible property type

##### **Contingency table for city**

```python
city_table = pd.crosstab(df['city'], df['type'], normalize='columns')
display(city_table)
```

Bar plot for city contingency table

```python
fig, axs = plt.subplots(1, 3, figsize=(25,10))
fig.subplots_adjust(wspace=0.5)

types = ['Condo', 'Multi-Family', 'Residential']

for i, prop_type in enumerate(types):
    city_table[prop_type].plot.barh(ax=axs[i])
    axs[i].set_xlim([0,1])
    axs[i].set_title(prop_type)
    axs[i].set_ylabel('city')
```

We can see from each of the bar plots above, the distributions of the type of properties in each city are somewhat different. From this, we can conclude that the city can potentially predict the type of property.

##### **Contingency table for state**

```python
state_table = pd.crosstab(df['state'], df['type'], normalize='columns')
display(state_table)
```

Bar plot for state contingency table

```python
fig, axs = plt.subplots(1, 3, figsize=(15,4))
fig.subplots_adjust(wspace=0.5)

types = ['Condo', 'Multi-Family', 'Residential']

for i, prop_type in enumerate(types):
    state_table[prop_type].plot.barh(ax=axs[i])
    axs[i].set_xlim([0,1])
    axs[i].set_title(prop_type)
    axs[i].set_ylabel('state')
```

Since the data is sampled from Sacremento, California there is only one state value for all of the data. Therefore, the state variable is not useful for predicting the type of property or any other target variable using this dataset.

##### **Contingency table for zip**

```python
zip_table = pd.crosstab(df['zip'], df['type'], normalize='columns')
display(zip_table)
```

Bar plot for zip contingency table

```python
fig, axs = plt.subplots(1, 3, figsize=(30,12))
fig.subplots_adjust(wspace=0.3)

types = ['Condo', 'Multi-Family', 'Residential']

for i, prop_type in enumerate(types):
    zip_table[prop_type].plot.barh(ax=axs[i])
    axs[i].set_xlim([0,1])
    axs[i].set_title(prop_type)
    axs[i].set_ylabel('zip')
    axs[i].set_yticklabels(axs[i].get_yticklabels(), fontsize=10)
```

We can see from each of the bar plots above, the distributions of the type of properties in each zip code are somewhat different. From this, we can conclude that the zip code can potentially predict the type of property, but most likely not very well.
